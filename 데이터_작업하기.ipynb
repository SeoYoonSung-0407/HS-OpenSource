```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/한국환경공단_전국 충전사업자별 전기차 충전기 구축 현황.csv",encoding='cp949')
df = df.drop(columns=['급속충전량','주소','충전소명','시설구분(대)','기종(대)','시설구분(소)','기종(소)','충전기타입','이용자제한'])
df = df.dropna(subset=['군구'])#기존 군구에서의 284개의 결측치 제거

# 시도별 충전기 개수 계산
chargers_by_region = df.groupby('시도')['충전기ID'].sum().reset_index()
chargers_by_region.rename(columns={'충전기ID': 'total_chargers'}, inplace=True)

# 전국 총 충전기 개수 계산
total_chargers = chargers_by_region['total_chargers'].sum()

# total_chargers를 DataFrame에 추가
total_row = pd.DataFrame({'시도': ['전체'], 'total_chargers': [total_chargers]})
chargers_with_total = pd.concat([chargers_by_region, total_row], ignore_index=True)

jeju_idx = chargers_with_total[chargers_with_total['시도'] == '충청북도'].index[0]
chargers_with_total.loc[jeju_idx + 1:] = chargers_with_total.loc[jeju_idx + 1:].shift(1)
chargers_with_total.loc[jeju_idx + 1] = {'시도': '전체', 'total_chargers': total_chargers}

chargers_with_total['total_chargers'] = chargers_with_total['total_chargers'].astype(int)

# 결과 확인
display(chargers_with_total)


file_path = "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/현재 전기차 충전기 수량 데이터.csv"

# CSV 파일로 저장
chargers_with_total.to_csv(file_path, index=False, encoding='utf-8-sig')

# 저장 완료 메시지
print(f"CSV 파일이 구글 드라이브에 저장되었습니다: {file_path}")
```

```python
import pandas as pd

# 연도별 파일 경로 리스트
file_paths = {
    2017: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2017_년도_충전기.csv",
    2018: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2018_년도_충전기.csv",
    2019: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2019_년도_충전기.csv",
    2020: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2020_년도_충전기.csv",
    2021: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2021_년도_충전기.csv",
    2022: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2022_년도_충전기.csv",
    2023: "/content/drive/MyDrive/[AI 웹 서비스 ]/미드 프로젝트/데이터 파일/2023_년도_충전기.csv"
}

# 결과 저장 리스트
yearly_totals = []

for year, path in file_paths.items():
    # 데이터 읽기
    df = pd.read_csv(path)

    # 충전기 개수 계산
    charging_stations_count = df.groupby(['설치년도', '시도', '군구', 'ID']).size().reset_index(name='충전기개수')

    # 시도별 충전기 총합 계산
    total_by_province = charging_stations_count.groupby('시도')['충전기개수'].sum().reset_index()

    # 연도 추가
    total_by_province['설치년도'] = year

    # 결과 저장
    yearly_totals.append(total_by_province)

# 모든 연도 결과 합치기
final_result = pd.concat(yearly_totals, ignore_index=True)

# 결과 확인
print(final_result)

# CSV로 저장
# final_result.to_csv("년도별_시도별_충전기_총합.csv", index=False, encoding='utf-8-sig')
# print("결과가 CSV로 저장되었습니다.")
```
